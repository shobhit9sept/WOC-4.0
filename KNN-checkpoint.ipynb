```python
training_data = train.values
testing_data = test.values
X_test = testing_data[:, 1:]
Y_test = testing_data[:, 0].reshape(testing_data.shape[0], 1)
X_train = training_data[:, 1:]
Y_train = training_data[:,0].reshape(training_data.shape[0], 1)
print("The shape of X_train is", X_train.shape)
print("The shape of X_test is", X_test.shape)
print("The shape of Y_train is", Y_train.shape)
print("The shape of Y_test is", Y_test.shape)
```

```python
prediction=np.zeros((14800,1))
for i in range(14800):
    diff_i = (X_train - X_test[i])**2
    sq_dist_i = np.sum(diff_i, axis = 1)
    dist_i = np.sqrt(sq_dist_i)
    sort_dist = np.argsort(dist_i)
    K_points = sort_dist[:5]
    counter = np.zeros((26,1))
    
    for j in K_points:
        index = Y_train[j]
        counter[index -1] += 1
    index=np.argmax(counter)
    prediction[i]=index+1
    print(prediction[i])
```

```python
count=0
for i in range(14800):
    if(prediction[i]==Y_test[i]):
        count+=1
(count/14800)*100
```
