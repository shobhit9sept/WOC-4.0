```python
training_data = train.values
testing_data = test.values
X_test = (testing_data[:, 1:])
Y_test = testing_data[:, 0].reshape(testing_data.shape[0],1)
X_train = (training_data[:, 1:])
Y_train = training_data[:, 0].reshape(training_data.shape[0],1)
print("The shape of X_train is", X_train.shape)
print("The shape of X_test is", X_test.shape)
print("The shape of Y_train is", Y_train.shape)
print("The shape of Y_test is", Y_test.shape)
m,n=X_train.shape
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def initial_parameters(n_x, n_h1, n_h2, n_y):
    theta_1 = np.random.randn(n_h1, n_x)*0.003

    
    theta_2 = np.random.randn(n_h2, n_y)*0.003

    
    return theta_1,theta_2
     
    
```

```python
def cost(theta1,theta2,X_train,Y_train):
    m,n=X_train.shape
    a1=X_train
    a2=sigmoid(a1@theta1.T)
    u=np.ones((m,1))
    a2=np.concatenate((u,a2),axis=1)
    A=sigmoid(a2@theta2.T)
    
    J=  -((1/m)*np.sum( Y*np.log(A) + (1-Y)*np.log(1-A)))
    return J


def sigmoidGradient(z):
 
  
    g = np.zeros(z.shape)
  
 
  
    g = sigmoid(z)*(1-sigmoid(z))
    return g
```

```python
theta1_diff=np.zeros((theta1.shape))
theta2_diff=np.zeros((theta2.shape))
```

```python
def back_prop(theta1,theta2,X_train,Y):
    m,n=X_train.shape
    alpha=0.05
    iterations = 200
    for i in range(iterations):
            a1=X_train
            a2=sigmoid(a1@theta1.T)
            u=np.ones((m,1))
            a2=np.concatenate((u,a2),axis=1)
            a3=sigmoid(a2@theta2.T)
            h_x = a3
    
            DELTA3 = a3 - Y
            m,n=(DELTA3@theta2).shape
            DELTA2 = (DELTA3@theta2) * (np.concatenate((np.ones((m,1)), sigmoidGradient(a1@theta1.T)), axis = 1))
            DELTA2 = DELTA2[:, 1:]
  
            theta1_grad = (1/m) * (DELTA2.T@a1)
            theta1 = theta1-alpha*theta1_grad
    
            theta2_grad = (1/m) * (DELTA3.T@a2)
            theta2 = theta2-alpha*theta2_grad
            J=cost(theta1,theta2,X_train,Y)
            if(i%(iterations/10) == 0):
                print("The value of our cost is: ", J)
    return theta1,theta2,h_x
```
